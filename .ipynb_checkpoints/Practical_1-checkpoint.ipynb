````python
# required libraries
import pandas as pd
import numpy as np
import datetime 
import os
import yfinance as yf
import matplotlib.pyplot as plt
````

````python
def fetch_stock_data(tickers, years=5, interval="1D"):
    end_date = pd.Timestamp.now()
    start_date = end_date - pd.DateOffset(years=years)
    
    os.makedirs("stock_data", exist_ok=True)
    for ticker_symbol in tickers:
        stock_data = yf.download(ticker_symbol, start=start_date, end=end_date, interval=interval)
        filename = f"stock_data/{ticker_symbol.split('.')[0]}_data.csv"
        stock_data.to_csv(filename, encoding="utf-8")
    
    print("Data has been downloaded successfully")
````

````python
tickers = ["NTPC.NS","HCLTECH.NS"]
fetch_stock_data(tickers)
````

````python
NTPC = pd.read_csv(os.path.join(os.getcwd(),"stock_data/NTPC_data.csv"),index_col="Date", parse_dates=True)
HCL = pd.read_csv(os.path.join(os.getcwd(),"stock_data/HCLTECH_data.csv"),index_col="Date", parse_dates=True)
````

````python
def normalize(column):
    min_val = column.min()
    max_val = column.max()
    return (column - min_val) / (max_val - min_val)
````

## NTPC analysis

````python
NTPC['Close'].resample("M").mean().plot.bar(figsize=(16,6))
````

````python
NTPC['Close'].resample("A").mean().plot.bar()
````

````python
title='NTPC Closing Stock Price '
ylabel='Closing Price'
xlabel='Closing Date'

ax = NTPC['Close'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)
````

````python
NTPC.rolling(window=7).mean().head(10)
````

````python
NTPC['Close'].plot(figsize=(12,5)).autoscale(axis='x',tight=True)
NTPC.rolling(window=30).mean()['Close'].plot()
````

````python
NTPC['Close: 30 Day Mean'] = NTPC['Close'].rolling(window=30).mean()
NTPC[['Close','Close: 30 Day Mean']].plot(figsize=(12,5)).autoscale(axis='x',tight=True)
````

````python
NTPC['Close'].expanding(min_periods=30).mean().plot(figsize=(12,5))
````

## Camparison Between NTPC & HCL

````python
plt.figure(figsize=(12, 6))
plt.plot(NTPC['Close'], label='NTPC')
plt.plot(HCL['Close'], label='HCL')
plt.title('Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
````

````python
# Calculate the correlation between NTPC and HCL
correlation = NTPC['Close'].corr(HCL['Close'])
print(f'Correlation between NTPC and HCL: {correlation:.2f}')
````

````python
#Shifts (e.g., shift one month forward)
NTPC['Close_Shifted'] = NTPC['Close'].shift(1)
HCL['Close_Shifted'] = HCL['Close'].shift(1)

NTPC['30D_Rolling'] = NTPC['Close'].rolling(window=30).mean()
HCL['30D_Rolling'] = HCL['Close'].rolling(window=30).mean()
````

````python
# Plotting rolling mean for both stocks
plt.figure(figsize=(12, 6))
plt.plot(NTPC['30D_Rolling'], label='NTPC 30-Day Rolling Mean')
plt.plot(HCL['30D_Rolling'], label='HCL 30-Day Rolling Mean')
plt.title('Stocks Rolling Means')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
````
